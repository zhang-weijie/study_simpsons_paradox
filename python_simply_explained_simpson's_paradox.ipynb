```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Department': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Male_Applicant': [825, 560, 325, 417, 191, 373],
    'Admission_Rate_Male': [0.62, 0.63, 0.37, 0.33, 0.28, 0.06],
    'Female_Applicant': [108, 25, 593, 375, 393, 341],
    'Admission_Rate_Female': [0.82, 0.68, 0.34, 0.35, 0.24, 0.07]
}

df = pd.DataFrame(data)
df.describe()

# 计算总体录取率 (Calculate overall admission rates)
total_male_applicants = df['Male_Applicant'].sum()
total_female_applicants = df['Female_Applicant'].sum()
total_admitted_male = (df['Male_Applicant'] * df['Admission_Rate_Male']).sum()
total_admitted_female = (df['Female_Applicant'] * df['Admission_Rate_Female']).sum()

overall_admission_rate_male = total_admitted_male / total_male_applicants
overall_admission_rate_female = total_admitted_female / total_female_applicants

# 绘制总体录取率柱状图 (Plot a bar chart for overall admission rates)
plt.bar(['Male', 'Female'], [overall_admission_rate_male, overall_admission_rate_female])
plt.title('Overall Admission Rate')  # 标题 (Title)
plt.xlabel('Gender')  # X轴标签 (X-axis Label)
plt.ylabel('Admission Rate')  # Y轴标签 (Y-axis Label)
plt.show()
```

```python
# Compare admission rates
# 比较录取率
fig, ax = plt.subplots()
bar_width = 0.35
index = np.arange(len(data['Department']))

# Plotting admission rates for males and females
# 绘制男女录取率
ax.bar(index, data['Admission_Rate_Male'], bar_width, label='Male Admission Rate')
ax.bar(index + bar_width, data['Admission_Rate_Female'], bar_width, label='Female Admission Rate')

ax.set_xlabel('Department')
ax.set_ylabel('Admission Rate')
ax.set_title('Admission Rates by Gender and Department')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(data['Department'])
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

data = {
    'Department': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Male_Applicant': [825, 560, 325, 417, 191, 373],
    'Female_Applicant': [108, 25, 593, 375, 393, 341]
}

# Calculate total applicants for each gender
total_male_applicants = sum(data['Male_Applicant'])
total_female_applicants = sum(data['Female_Applicant'])

# Calculate percentage of applicants in each department for each gender
male_percentage = [x / total_male_applicants * 100 for x in data['Male_Applicant']]
female_percentage = [x / total_female_applicants * 100 for x in data['Female_Applicant']]

# Plotting pie charts for male and female applicants in each department
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for male applicants
ax1.pie(male_percentage, labels=data['Department'], autopct='%1.1f%%', startangle=140)
ax1.set_title('Male Applicants Distribution by Department')

# Pie chart for female applicants
ax2.pie(female_percentage, labels=data['Department'], autopct='%1.1f%%', startangle=140)
ax2.set_title('Female Applicants Distribution by Department')

plt.show()

# Preference Analysis
# 偏好分析
fig, ax = plt.subplots()

# Calculating and plotting the proportion of male and female applicants in each department
# 计算并绘制每个部门男女申请人的比例
male_proportion = [m / (m + f) for m, f in zip(data['Male_Applicant'], data['Female_Applicant'])]
female_proportion = [f / (m + f) for m, f in zip(data['Male_Applicant'], data['Female_Applicant'])]

ax.bar(index, male_proportion, bar_width, label='Male Proportion')
ax.bar(index + bar_width, female_proportion, bar_width, label='Female Proportion')

ax.set_xlabel('Department')
ax.set_ylabel('Proportion')
ax.set_title('Gender Proportions by Department')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(data['Department'])
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data dictionary
data = {
    'Department': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Male_Applicant': [825, 560, 325, 417, 191, 373],
    'Admission_Rate_Male': [0.62, 0.63, 0.37, 0.33, 0.28, 0.06],
    'Female_Applicant': [108, 25, 593, 375, 393, 341],
    'Admission_Rate_Female': [0.82, 0.68, 0.34, 0.35, 0.24, 0.07]
}

# Extracting data from the dictionary
departments = data['Department']
male_applicants = data['Male_Applicant']
male_admission_rates = data['Admission_Rate_Male']
female_applicants = data['Female_Applicant']
female_admission_rates = data['Admission_Rate_Female']

# Calculate the number of male and female admissions in each department
male_admitted = [rate * applicants for rate, applicants in zip(male_admission_rates, male_applicants)]
female_admitted = [rate * applicants for rate, applicants in zip(female_admission_rates, female_applicants)]

# Create a Figure and an Axes
fig, ax = plt.subplots()

# Plot male vectors
ax.quiver(np.zeros(len(departments)), np.zeros(len(departments)), male_applicants, male_admitted,
          angles='xy', scale_units='xy', scale=1, color='r', label='Male')

# Plot female vectors
ax.quiver(np.zeros(len(departments)), np.zeros(len(departments)), female_applicants, female_admitted,
          angles='xy', scale_units='xy', scale=1, color='b', label='Female')

# Set the axis limits
ax.set_xlim([0, 1000])
ax.set_ylim([0, 1000])

# Add axis labels
ax.set_xlabel('Total Applicants')
ax.set_ylabel('Admitted Applicants')

# Add a grid
ax.grid(True)

# Add department labels
for i, department in enumerate(departments):
    ax.text(male_applicants[i], male_admitted[i], department, ha='right', va='bottom', color='r')
    ax.text(female_applicants[i], female_admitted[i], department, ha='left', va='top', color='b')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```
